```python
temp = pd.concat([df_new['Negative_Review'][:10000], df_new['Positive_Review'][:10000]],axis=0).reset_index()
```

```python
temp = temp.drop(['index'],axis=1)
```

```python
temp['sentiment'] = 0
```

```python
temp.to_csv('processed_data.csv')
```

```python
data = temp.copy()
```

```python
import re
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```python
ps = PorterStemmer()

for i in range(len(data)):
    data['review'][i] = re.sub('[^a-zA-Z]', " ",data['review'][i])
    
    sent = data['review'][i].split()
    
    sent = [ps.stem(word) for word in sent if word not in stopwords.words('english')]
    
    sent = ' '.join(sent)
    
    data['review'][i] = sent
    
```

```python
X = data['review']
y = data['sentiment']
```

# Word Embedding

```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

#### One hot

```python
voc_size = 50000

onehot_repr = [ one_hot(word,voc_size) for word in X]
onehot_repr
```

```python
sent_length = 1000
embedded_docs = pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
embedded_docs
```

# Creating model

```python
from tensorflow.keras.layers import LSTM, Dense, Dropout, Embedding
from tensorflow.keras.models import Sequential
```

```python
embedding_vector_featres = 64

model = Sequential()
model.add(Embedding(voc_size,embedding_vector_featres,input_length=sent_length))
model.add(Dropout(0.4))
model.add(LSTM(128))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

model.summary()
```

```python
import numpy as np

X_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_final,y_final,test_size=0.2,random_state=42)
```

# Model Tranining

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test), epochs=2, batch_size=32)
```

```python
from sklearn.metrics import confusion_matrix

y_pred = model.predict_classes(X_test)

confusion_matrix(y_test,y_pred)
```
